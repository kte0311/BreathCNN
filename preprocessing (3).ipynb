```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
train = pd.read_csv('/content/drive/MyDrive/머신러닝_kaggle2/train.csv')
test = pd.read_csv('/content/drive/MyDrive/머신러닝_kaggle2/test.csv')

# 저장할 디렉토리 경로 설정
base_drive_path = "/content/drive/MyDrive/머신러닝_kaggle2"

train_log_mel_dir = os.path.join(base_drive_path, "train_log_mel")
test_log_mel_dir = os.path.join(base_drive_path, "test_log_mel")

# 디렉토리 생성 (없는 경우)
os.makedirs(train_log_mel_dir, exist_ok=True)
os.makedirs(test_log_mel_dir, exist_ok=True)
```

```python
for i in train['ID'] :
  j = i.replace("_I_", '_')
  j = j.replace("_E_", '_')
  path = f"/content/drive/MyDrive/머신러닝_kaggle2/train/{j}.wav"
  y, sr = librosa.load(path)

  n_fft = 2048       # FFT 윈도우 크기
  hop_length = 512   # 프레임 이동 간격
  n_mels = 128

  # Log Mel Spectrogram 계산
  S_stft = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
  S_magnitude = np.abs(S_stft)
  mel_spectrogram = librosa.feature.melspectrogram(S=S_magnitude, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
  log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)
  np.save(os.path.join(train_log_mel_dir, i), log_mel_spectrogram)
```

```python
for i in test['ID'] :
  j = i.replace("_I_", '_')
  j = j.replace("_E_", '_')
  path = f"/content/drive/MyDrive/머신러닝_kaggle2/test/{j}"
  y, sr = librosa.load(path)

  n_fft = 2048
  hop_length = 512
  n_mels = 128

  # Log Mel Spectrogram 계산
  S_stft = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
  S_magnitude = np.abs(S_stft)
  mel_spectrogram = librosa.feature.melspectrogram(S=S_magnitude, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
  log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)
  np.save(os.path.join(test_log_mel_dir, i), log_mel_spectrogram)
```
